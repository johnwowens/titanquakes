```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
from matplotlib.ticker import MultipleLocator, EngFormatter
pd.options.plotting.backend = "plotly"

df = pd.read_csv('swavearrivals.csv')
```

```python
print(df)
```

```python
dist = df['Distance from source (actual, km)']
tSest = df['tS (est)']
tSTauP = df['tS (TauP)']
tSvis = df['tS (visual)']
fig, ax = plt.subplots()
markers = ['o','s','D']
labels=['Estimated', 'TauP', 'Visual']
ind = 0
for item in [tSest, tSTauP, tSvis]:
    ax.plot(dist, item, marker=markers[ind], linestyle='', ms=6, label=labels[ind])
    ind += 1
fig.set_size_inches(10,6)
plt.xlim(350,1650)
# plt.ylim(0.0000001, 0.001)
ax.legend(fontsize=14)
# plt.gca().invert_xaxis()
ax.set_ylabel('S-wave arrival time (s)', fontsize=16)
ax.set_xlabel('Epicentral distance (km)', fontsize=16)
ax.tick_params(axis='both', which='major', labelsize=12)
# ax.yaxis.set_major_formatter(EngFormatter(unit=u"°"))
# ax.xaxis.set_major_formatter(EngFormatter(unit=u"°"))
ax.minorticks_on()
#plt.savefig('swavearrivals.png',dpi=500)
```
