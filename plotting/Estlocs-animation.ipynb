```python
import matplotlib.pyplot as plt
import matplotlib.animation as animation
import numpy as np
import pandas as pd
from matplotlib.ticker import MultipleLocator, EngFormatter
pd.options.plotting.backend = "plotly"

df = pd.read_csv('estlocs.csv')
```

```python
df2=df
groups = df2.groupby('delta')
fig, ax = plt.subplots()
ax.margins(0.05) # Optional, just adds 5% padding to the autoscaling
markers = ['o','x','s','*']
dip = ['2','5','9','Actual']
marker_ind=0

ims = []

for name, group in groups:
    if marker_ind <= 2:
        #ax.plot(group['Long-est2'], group['Lat-est2'],\
        #        marker=markers[marker_ind], fillstyle='none', linestyle='', ms=8, label='δ='+dip[marker_ind]+'$^\circ$')
        im = plt.imshow(ax.plot(group['Long-est2'], group['Lat-est2'],\
                marker=markers[marker_ind], fillstyle='none', linestyle='', ms=8, label='δ='+dip[marker_ind]+'$^\circ$'), animated=True)
        ims.append([im])
    else:
        #ax.plot(group['Long-est2'], group['Lat-est2'],\
        #        marker=markers[marker_ind], fillstyle='none', linestyle='', ms=12, label=dip[marker_ind])
        im = plt.imshow(ax.plot(group['Long-est2'], group['Lat-est2'],\
                marker=markers[marker_ind], fillstyle='none', linestyle='', ms=12, label=dip[marker_ind]), animated=True)
        ims.append([im])
    marker_ind += 1
fig.set_size_inches(8, 6)
plt.xlim(202,192)
plt.ylim(-20, 20)
ax.legend(fontsize=16)
#plt.gca().invert_xaxis()
ax.set_ylabel('Latitude', fontsize=16)
ax.set_xlabel('Longitude', fontsize=16)
ax.tick_params(axis='both', which='major', labelsize=12)
ax.yaxis.set_major_formatter(EngFormatter(unit=u"°"))
ax.xaxis.set_major_formatter(EngFormatter(unit=u"°"))
ax.minorticks_on()
# plt.savefig('estlocs.png',dpi=1000)

ani = animation.ArtistAnimation(fig, ims, interval=1000, blit=True,
                                repeat_delay=3000)
```
